
*   **Enter the string input using lower letters and the string result in capital letters with X resembling a letter that doesn't exist, G resembling a letter in the correct spot, and Y resembling a letter that exists but is in the wrong spot**

*   whenever using a method in python, create copies of the parameters(including dataframes), rather than just editing the dataframe itself
*   use mask to create a boolean series of indexes and use the .loc function in python to filter out the dataframes that you don't wan
*   iloc should be used when accessing row/col of dataframe using index and loc should be used when accessing the names of rows/columns















```python
import pandas as pd
df = pd.read_csv("5_letters.csv")
df
```

```python
def filterWords(df, inputString, inputResult):
  for i in range(5):
    if (inputResult[i] != 'X') & (inputResult[i] != 'G') & (inputResult[i] != 'Y'):
      print(inputResult[i])
      return "Invalid input"
  df_updated = filterGreenLetters(df, inputString, inputResult)
  df_updated = filterGrayLetters(df_updated, inputString, inputResult)
  df_updated = filterYellowLetters(df_updated, inputString, inputResult)
  if df_updated.size == 0:
    return "It seems like you have given me incorrect input somewhere"
  if df_updated.size == 5:
    toReturn = "Congratulations, your word is: "
    for i in range(5):
      toReturn += df_updated.iloc[0, i]
    return toReturn
  return (df_updated)
```

```python
def filterGreenLetters(df,  inputString, inputResult):
  df_updated = df.copy()  # Make a copy of the DataFrame
  for i in range(5):
    if inputResult[i] == 'G':
      mask = df_updated.iloc[:, i] == inputString[i]  # Filter using column name
      df_updated = df_updated.loc[mask]
  return (df_updated)
```

```python
def filterGrayLetters(df, inputString, inputResult):
  df_updated = df.copy()  # Make a copy of the DataFrame
  for i in range(5):
    if inputResult[i] == 'X':
        for j in range(5):
          mask = df_updated.iloc[:, j] != inputString[i]  # Filter using column name
          df_updated = df_updated.loc[mask]
  return (df_updated)
```

```python
def filterYellowLetters(df, inputString, inputResult):
  df_updated = df.copy()  # Make a copy of the DataFrame
  for i in range(5):
    if inputResult[i] == 'Y':
      mask = df_updated.iloc[:, i] != inputString[i]  # Filter using column name by creating boolean series
      df_updated = df_updated.loc[mask]
      if (i == 0): #have or statement
        mask = (df_updated.iloc[:, 1] == inputString[i]) | (df_updated.iloc[:, 2] == inputString[i]) | (df_updated.iloc[:, 3] == inputString[i]) | (df_updated.iloc[:, 4] == inputString[i]) # Filter using column name
      elif (i == 1):
        mask = (df_updated.iloc[:, 0] == inputString[i]) | (df_updated.iloc[:, 2] == inputString[i]) | (df_updated.iloc[:, 3] == inputString[i]) | (df_updated.iloc[:, 4] == inputString[i]) # Filter using column name
      elif (i == 2):
        mask = (df_updated.iloc[:, 0] == inputString[i]) | (df_updated.iloc[:, 1] == inputString[i]) | (df_updated.iloc[:, 3] == inputString[i]) | (df_updated.iloc[:, 4] == inputString[i])
      elif (i == 3):
        mask = (df_updated.iloc[:, 0] == inputString[i]) | (df_updated.iloc[:, 1] == inputString[i]) | (df_updated.iloc[:, 2] == inputString[i]) | (df_updated.iloc[:, 4] == inputString[i])
      elif (i == 4):
        mask = (df_updated.iloc[:, 0] == inputString[i]) | (df_updated.iloc[:, 1] == inputString[i]) | (df_updated.iloc[:, 2] == inputString[i]) | (df_updated.iloc[:, 3] == inputString[i])
      df_updated = df_updated.loc[mask]
  return (df_updated)
```
